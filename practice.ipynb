## Merge operations

Merge operations allow you to combine data from two or more DataFrames based on common columns (or keys). These operations are similar to SQL joins. Let me explain the key concepts:

### Types of Merge Operations:
<b>Inner Merge (Intersection)</b>: Only the rows with matching keys in both DataFrames are included in the result. Rows with non-matching keys are excluded.

<b>Outer Merge (Union)</b>: All rows from both DataFrames are included in the result. Missing values are filled with NaN.

<b>Left Merge (Left Outer Join)</b>: All rows from the left DataFrame are included, and matching rows from the right DataFrame are added. Non-matching rows from the right DataFrame are excluded.


<b>Right Merge (Right Outer Join)</b>: Similar to left merge, but all rows from the right DataFrame are included.


### usage

merged_df = pd.merge(left_df, right_df, on='key_column', how='merge_type')
