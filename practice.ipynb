Creating dataframe from the list

```python
employees_list = [
    ["Alice", 70000, 8400, "HR"],
    ["Bob", 85000, 10200, "Finance"],
    ["Charlie", 56000, 6720, "Engineering"],
    ["David", 75000, 9000, "Marketing"],
    ["Eve", 94000, 11280, "IT"],
    ["Frank", 62000, 7440, "Sales"],
    ["Grace", 81000, 9720, "Operations"],
    ["Hank", 53000, 6360, "Legal"],
    ["Ivy", 68000, 8160, "Customer Service"],
    ["Jack", 91000, 10920, "Research"]
]
df=pd.DataFrame(data=employees_list,columns=['Name','Income','PF','Departmrnt'])
df
```

### <b>df.iloc</b> is a method in the Pandas library used for integer-location-based indexing to select rows and columns by their integer position. It is a powerful and flexible way to access data within a DataFrame, and it supports both single and multi-axis indexing

```python
#secting single row by its index positio
df.iloc[0]
```

```python
# selecting multiple rows by iindex position
df.iloc[0:5]
```

```python
# Selecting multiple rows and columns by their index positions:
# Select the first three rows and the first two columns

df.iloc[0:3,0:2]
```

```python
# selecting entire rows and columns
df.iloc[:3]
```

### <b>df.loc</b> is a property of a pandas DataFrame that is used to access a group of rows and columns by labels or a boolean array. The loc method allows for label-based indexing and is useful for selecting specific rows and columns, updating values, and filtering data.

```python
# filtering
df.loc[df['PF']>8000]
```

```python
df
```

```python
# updating the values
df.loc[df['Name']=='Alice','PF']=9000
```
